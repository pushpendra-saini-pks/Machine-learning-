### Independent and dependent features 

```python
x=dataset.data 
y=dataset.target 
```

### Train test split 

```python
from sklearn.model_selection import train_test_split 
x_train ,x_test ,y_train , y_test = train_test_split(x,y,test_size = 0.33,random_state = 42) 
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier = DecisionTreeClassifier()
```

```python
treeclassifier 
```

```python
treeclassifier.fit(x_train , y_train )
```

```python
import matplotlib.pyplot as plt 
```

```python
from sklearn import tree 
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled = True )
```

## Post Pruning 

- post - pruning , also known as "backward pruning " involves growing a full decision tree first and then removing or collapsing modes in a bottom - up manner .
- Once the tree is fully grown , pruning techniques such as cost complexity pruning (also known as minimal cost commplexity pruning or weakest link pruning ) are applind . This involves evaluating the impact of removing a subtree (or a single node ) on the overall performance of the tree .
- Pruning decisions are typically guided by a pruning criterion , such as cross - validation error , to determine the optimal level of complexity for the tree.
- Post - pruning can be computationally more expensive than pre-pruning since it involves growing a full tree before pruning .

###  Post Pruning with max depth = 2

```python
treeclassifier = DecisionTreeClassifier(max_depth = 2)
treeclassifier.fit(x_train , y_train )
```

```python
from sklearn import tree 
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled = True )
```

```python
# In terms of entropy 
treeclassifier = DecisionTreeClassifier(criterion='entropy',max_depth = 2)
treeclassifier.fit(x_train , y_train )
```

```python
from sklearn import tree 
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled = True )
```

```python
y_pred = treeclassifier.predict(x_test)
```

```python
y_pred 
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score , classification_report 
```

```python
print (accuracy_score(y_test , y_pred))
print (classification_report (y_test , y_pred))
```

### Pre-pruning / Hyperparameter Tuning for large dataset 

- This technique is used before construction of decision tree. 
- Pre-Pruning can be done using Hyperparameter tuning
- Overcome the overfitting issue- 
In this blog i will use GridSearchCV for Hyperparameter tunin
- by imposing these constraints during the tree construction process.pre-pruning prevents the tree from becoming overly complex and captures the most important in the data without memorizing noise or outliers ..ning.

```python
parameter = {'criterion':['gini','entropy','log_loss'],
            'splitter':['best','random'],
            'max_depth':[1,2,3,4,5],
            'max_features':['auto','sqrt','log2']}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf = GridSearchCV(DecisionTreeClassifier(),param_grid=parameter , cv=5 , scoring = 'accuracy')
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
clf.fit(x_train , y_train)
```

```python
clf.best_params_
```

```python
y_pred_test = clf.predict(x_test)
```

```python
print (accuracy_score(y_test , y_pred_test))
print (classification_report(y_test , y_pred_test))
```

**In summary , pre-pruning sets constraints during the tree construction process to prevent overfitting , while post - pruning involves growing a full tree first and then removing unnecessary nodes to improve generalization performance . both techniques aim to strike a balance between model complexity and predictive accuracy . the choice between pre- pruning and post- pruning often depends on the specific charateristics of the dataset and the computational resources available.**
