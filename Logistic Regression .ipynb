## Logistic Regression Implementation 

### Binary Classification 

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd 
```

```python
dataset=load_breast_cancer()
```

```python
dataset
```

```python
print(dataset.DESCR)
```

```python
dataset.keys()
```

```python
dataset.data
```

```python
df=pd.DataFrame(dataset.data,columns = dataset.feature_names)
```

```python
df
```

```python
df['target']=dataset.target
```

```python
df.head()
```

```python
df.isnull().sum()
```

### Independent and Dependent features

```python
x=df.iloc[:,:-1]
x
```

```python
y=df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split  
```

```python
x_train , x_test, y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train 
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier
```

```python
classifier.fit(x_train,y_train)
```

```python
 classifier.predict_proba(x_test)
```

```python
y_pred = classifier.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

### Confusion matrix , accuracy score  , classification report 

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
print (confusion_matrix(y_test,y_pred))
print (accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

### Hyperparameter Tuning 

#### Grid search CV

```python
from sklearn.model_selection import GridSearchCV
import warnings 
warnings.filterwarnings('ignore')
```

```python
parameters = {'penalty':('l1','l2','elaticnet',None), 'C':[1,10,20]}
```

```python
clgs = GridSearchCV(classifier,param_grid = parameters , cv=5)
```

### split train data to validation data 

```python
clgs.fit(x_train ,y_train)
```

```python
clgs.best_params_
```

```python
classifier = LogisticRegression(C=20,penalty = 'l2')
```

```python
classifier.fit(x_train , y_train )
```

```python
y_pred = classifier.predict(x_test)
```

```python
print (confusion_matrix(y_test,y_pred))
print (accuracy_score(y_test,y_pred))
print (classification_report(y_test,y_pred))
```

### Randomized Search CV

```python
from sklearn.model_selection import RandomizedSearchCV 
```

```python
random_cv = RandomizedSearchCV(LogisticRegression(),param_distributions=parameters , cv=5)
```

```python
random_cv.fit(x_train , y_train)
```
